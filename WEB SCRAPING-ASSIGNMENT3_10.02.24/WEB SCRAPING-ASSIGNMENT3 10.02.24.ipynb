Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“. 


```python
br_name = []
pr_name = []
price = []
exchange = []
exp_del = []
avail = []
url = []
start = 0
end = 3

for page in range(start,end):
    brand = driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
    for b in brand:
        br_name.append(b.text[:7])
    
    name = driver.find_elements(By.XPATH,'//span[@class="a-size-base-plus a-color-base a-text-normal"]')
    for b in brand:
        pr_name.append(b.text)
    
    price1 = driver.find_elements(By.XPATH,'//span[@class="a-price-whole"]')
    for p in price1:
        price.append(p.text)
    
    exp_delivery = driver.find_elements(By.XPATH,'//span[@class="a-color-base a-text-bold"]')
    for e in exp_delivery:
        exp_del.append(e.text)

    purl = driver.find_elements(By.XPATH,'.//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for ur in purl:
        url.append(ur.get_attribute('href'))
        
data = {'Brand':br_name[:100],
       'Product_Name:':pr_name[:100],
       'Price':price[:100],
       'Expected_Delivery':exp_del[:100],
       'Product_URL':url[:100]}

df = pd.DataFrame(data)
df
```

**9. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in “London” location.**
You have to scrape hostel name, distance from city centre, ratings, total reviews, overall reviews, privates from price, dorms from price

```python
driver = webdriver.Chrome()
driver.get('https://www.hostelworld.com/pwa/wds/s?q=London,%20England&country=London&city=London&type=city&id=3&from=2024-02-11&to=2024-02-14&guests=2&page=1')
```

```python
hname = []
hdist = []
ratings = []
review= []
pprice = []
dprice = []
start = 0
end = 4

for page in range(start,end):

    hna = driver.find_elements(By.XPATH,'//div[@class="property-name"]')
    for h in hna:
        hname.append(h.text)

    hd = driver.find_elements(By.XPATH,'//span[@class="distance-description"]')
    for d in hd:
        hdist.append(d.text)

    rat = driver.find_elements(By.XPATH,'//span[@class="number"]')
    for r in rat:
        ratings.append(r.text)

    rev = driver.find_elements(By.XPATH,'//span[@class="keyword"]')
    for rv in rev:
        review.append(rv.text)

    pp = driver.find_elements(By.XPATH,'//div[@class="accommodation-price"]/span')
    for p in pp:
        pprice.append(p.text)

    dp = driver.find_elements(By.XPATH,'//div[@class="accommodation-price"]/strong')
    for p1 in dp:
        dprice.append(p1.text)

data = { 'Hostel Name':hname[:20],
       'Hostel Distance':hdist[:20],
       'Ratings':ratings[:20],
       'Review':review[:20],
       'Private Price':pprice[:20],
       'Dorm Price':dprice[:20]}

df = pd.DataFrame(data)
df
```
