```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(page.content)
header = []

for i in soup.find_all('span',class_ = "mw-headline"):
    header.append(i.text)
    
df = pd.DataFrame(header, columns = ['Header'])
df
```

**2) Write s python program to display list of respected former presidents of Indiafrom https://presidentofindia.nic.in/former-presidents and make data frame..**

```python
page = requests.get('https://presidentofindia.nic.in/former-presidents')
soup = BeautifulSoup(page.content)

names = []
rank = []

for i in soup.find_all('h3'):
    names.append(i.text)

for r in soup.find_all('h5'):
    rank.append(r.text)

data = {'Name of President': names,
       'Rank': rank}

df = pd.DataFrame(data)
print(df)
```

**Q 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-**

i) Headline

ii) Time

iii) News Link 

```python
page = requests.get("https://www.cnbc.com/world/?region=world")
soup = BeautifulSoup(page.content)
headlines = []
times = []
links = []

for h in soup.find_all('div', class_ = 'RiverPlusCard-container'):
    headlines.append(h.text)

for t in soup.find_all('span', class_ = 'RiverByline-datePublished'):
    times.append(t.text)

for n in soup.find_all('a', href = True):
    links.append(n['href'])

data = {'Headline': headlines,
       'Time': times,
       'News Link': links}

data
```

i) Paper Title
ii) Authors

```python

page = requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
soup = BeautifulSoup(page.content) 
    
paper = []
authors = []
date = []
url = []
    
for p in soup.find_all('h2', class_ = 'sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg'):
    paper.append(p.text)
    
for a in soup.find_all('span', class_= 'sc-1w3fpd7-0 dnCnAO'):
    authors.append(a.text)
    
for d in soup.find_all('span',class_ = 'sc-1thf9ly-2 dvggWt'):
    date.append(d.text)

for u in soup.find_all('a', href = True):
    url.append(u['href'])

data ={'Paper Title':paper, 'Authors': authors, 
 'Published Date': date, 'Paper URL': url}

data
```

**Q 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame-**

i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page

soup = BeautifulSoup(page.content)
soup

rest_names = []
Cuisines = []
Location = []
Ratings = []
Image_URL = []
    
def showrestdata():
    
    for i in soup.find_all('div', class_ = 'restnt-info cursor'):
        rest_names.append(i.text)
    
    for c in soup.find_all('span', class_ = 'double-line-ellipsis'):
        Cuisines.append(c.text[25:])

    for l in soup.find_all('div', class_='restnt-loc ellipsis'):
        Location.append(l.text)

    for r in soup.find_all('div', class_='restnt-rating rating-4'):
        Ratings.append(r.text)

    for u in soup.find_all('img', class_='no-img'):
        Image_URL.append(u['data-src'])
        
showrestdata()

data = {
  'Restaurant Name': rest_names,
  'Cuisine': Cuisines,
  'Location': Location,
  'Ratings': Ratings,
  'Image URL': Image_URL
}

df = pd.DataFrame(data)
print(df)

```
